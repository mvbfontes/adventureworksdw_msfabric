##### Criando e carregando a tabela "bronze_FactInternetSales"


Abrir o arquivo CSV utilizando Notepad++ para descobrir codificação

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, TimestampType, DecimalType

schema_stage = StructType([
StructField("ProductKey", StringType(), True),
StructField("OrderDateKey", StringType(), True),
StructField("DueDateKey", StringType(), True),
StructField("ShipDateKey", StringType(), True),
StructField("CustomerKey", StringType(), True),
StructField("PromotionKey", StringType(), True),
StructField("CurrencyKey", StringType(), True),
StructField("SalesTerritoryKey", StringType(), True),
StructField("SalesOrderNumber", StringType(), True),
StructField("SalesOrderLineNumber", StringType(), True),
StructField("RevisionNumber", StringType(), True),
StructField("OrderQuantity", StringType(), True),
StructField("UnitPrice", StringType(), True),
StructField("ExtendedAmount", StringType(), True),
StructField("UnitPriceDiscountPct", StringType(), True),
StructField("DiscountAmount", StringType(), True),
StructField("ProductStandardCost", StringType(), True),
StructField("TotalProductCost", StringType(), True),
StructField("SalesAmount", StringType(), True),
StructField("TaxAmt", StringType(), True),
StructField("Freight", StringType(), True),
StructField("CarrierTrackingNumber", StringType(), True),
StructField("CustomerPONumber", StringType(), True),
StructField("OrderDate", StringType(), True),
StructField("DueDate", StringType(), True),
StructField("ShipDate", StringType(), True)
])

df = spark.read.format("csv").load("Files/FactInternetSales.csv",header='false',sep='|',schema=schema_stage,encoding="UTF-16")

df.writeTo("bronze_factinternetsales").createOrReplace()

```

##### Criando a carregando a tabela "silver_factinternetsales"

```python
from pyspark.sql.functions import col
from pyspark.sql.types import StringType,BooleanType,DateType,IntegerType,DecimalType,TimestampType

df_convert = df.withColumn("ProductKey",col("ProductKey").cast(IntegerType())) \
.withColumn("OrderDateKey",col("OrderDateKey").cast(IntegerType())) \
.withColumn("DueDateKey",col("DueDateKey").cast(IntegerType())) \
.withColumn("ShipDateKey",col("ShipDateKey").cast(IntegerType())) \
.withColumn("CustomerKey",col("CustomerKey").cast(IntegerType())) \
.withColumn("PromotionKey",col("PromotionKey").cast(IntegerType())) \
.withColumn("CurrencyKey",col("CurrencyKey").cast(IntegerType())) \
.withColumn("SalesTerritoryKey",col("SalesTerritoryKey").cast(IntegerType())) \
.withColumn("SalesOrderNumber",col("SalesOrderNumber").cast(StringType())) \
.withColumn("SalesOrderLineNumber",col("SalesOrderLineNumber").cast(IntegerType())) \
.withColumn("RevisionNumber",col("RevisionNumber").cast(IntegerType())) \
.withColumn("OrderQuantity",col("OrderQuantity").cast(IntegerType())) \
.withColumn("UnitPrice",col("UnitPrice").cast(DecimalType())) \
.withColumn("ExtendedAmount",col("ExtendedAmount").cast(DecimalType())) \
.withColumn("UnitPriceDiscountPct",col("UnitPriceDiscountPct").cast(DecimalType())) \
.withColumn("DiscountAmount",col("DiscountAmount").cast(DecimalType())) \
.withColumn("ProductStandardCost",col("ProductStandardCost").cast(DecimalType())) \
.withColumn("TotalProductCost",col("TotalProductCost").cast(DecimalType())) \
.withColumn("SalesAmount",col("SalesAmount").cast(DecimalType())) \
.withColumn("TaxAmt",col("TaxAmt").cast(DecimalType())) \
.withColumn("Freight",col("Freight").cast(DecimalType())) \
.withColumn("CarrierTrackingNumber",col("CarrierTrackingNumber").cast(StringType())) \
.withColumn("CustomerPONumber",col("CustomerPONumber").cast(StringType())) \
.withColumn("OrderDate",col("OrderDate").cast(TimestampType())) \
.withColumn("DueDate",col("DueDate").cast(TimestampType())) \
.withColumn("ShipDate",col("ShipDate").cast(TimestampType())) 


df_convert.printSchema

df_convert.writeTo("silver_factinternetsales").createOrReplace()
```
