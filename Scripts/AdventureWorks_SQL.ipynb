##### Criando a carregando a tabela "bronze_factinternetsales" (Python)

Abrir arquivo CSV no Notepad++ para identificar codificação

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, TimestampType, DecimalType


schema_stage = StructType([
StructField("ProductKey", StringType(), True),
StructField("OrderDateKey", StringType(), True),
StructField("DueDateKey", StringType(), True),
StructField("ShipDateKey", StringType(), True),
StructField("CustomerKey", StringType(), True),
StructField("PromotionKey", StringType(), True),
StructField("CurrencyKey", StringType(), True),
StructField("SalesTerritoryKey", StringType(), True),
StructField("SalesOrderNumber", StringType(), True),
StructField("SalesOrderLineNumber", StringType(), True),
StructField("RevisionNumber", StringType(), True),
StructField("OrderQuantity", StringType(), True),
StructField("UnitPrice", StringType(), True),
StructField("ExtendedAmount", StringType(), True),
StructField("UnitPriceDiscountPct", StringType(), True),
StructField("DiscountAmount", StringType(), True),
StructField("ProductStandardCost", StringType(), True),
StructField("TotalProductCost", StringType(), True),
StructField("SalesAmount", StringType(), True),
StructField("TaxAmt", StringType(), True),
StructField("Freight", StringType(), True),
StructField("CarrierTrackingNumber", StringType(), True),
StructField("CustomerPONumber", StringType(), True),
StructField("OrderDate", StringType(), True),
StructField("DueDate", StringType(), True),
StructField("ShipDate", StringType(), True)
])

df = spark.read.format("csv").load("Files/FactInternetSales.csv",header='false',sep='|',schema=schema_stage,encoding="UTF-16")

df.write.format("delta").mode("overwrite").saveAsTable("bronze_factinternetsales_sql")
```

##### Criando view "silver_factinternetsales_sql_new" para novo arquivo com dados de 2015

```python
%%sql

CREATE OR REPLACE TEMP VIEW vw_silver_factinternetsales_sql_new AS

SELECT
CAST(ProductKey AS INT) AS ProductKey,
CAST(OrderDateKey AS INT) AS OrderDateKey,
CAST(DueDateKey AS INT) AS DueDateKey,
CAST(ShipDateKey AS INT) AS ShipDateKey,
CAST(CustomerKey AS INT) AS CustomerKey,
CAST(PromotionKey AS INT) AS PromotionKey,
CAST(CurrencyKey AS INT) AS CurrencyKey,
CAST(SalesTerritoryKey AS INT) AS SalesTerritoryKey,
CAST(SalesOrderNumber AS VARCHAR(20)) AS SalesOrderNumber,
CAST(SalesOrderLineNumber AS INT) AS SalesOrderLineNumber,
CAST(RevisionNumber AS INT) AS RevisionNumber,
CAST(OrderQuantity AS INT) AS OrderQuantity,
CAST(UnitPrice AS DECIMAL(10,6)) AS UnitPrice,
CAST(ExtendedAmount AS DECIMAL(10,6)) AS ExtendedAmount,
CAST(UnitPriceDiscountPct AS FLOAT) AS UnitPriceDiscountPct,
CAST(DiscountAmount AS FLOAT) AS DiscountAmount,
CAST(ProductStandardCost AS DECIMAL(10,6)) AS ProductStandardCost,
CAST(TotalProductCost AS DECIMAL(10,6)) AS TotalProductCost,
CAST(SalesAmount AS DECIMAL(10,6)) AS SalesAmount,
CAST(TaxAmt AS DECIMAL(10,6)) AS TaxAmt,
CAST(Freight AS DECIMAL(10,6)) AS Freight,
CAST(CarrierTrackingNumber AS VARCHAR(25)) AS CarrierTrackingNumber,
CAST(CustomerPONumber AS VARCHAR(25)) AS CustomerPONumber,
CAST(OrderDate AS TIMESTAMP) AS OrderDate,
CAST(DueDate AS TIMESTAMP) AS DueDate,
CAST(ShipDate AS TIMESTAMP) AS ShipDate
FROM Sales.bronze_factinternetsales2015_sql
WHERE CAST(ProductKey AS INT) IS NOT NULL;
```

##### Realizando UPSERT dos novos dados na tabela existente

```python
%%sql

MERGE INTO silver_factinternetsales_sql
USING vw_silver_factinternetsales_sql_new
ON 
silver_factinternetsales_sql.ProductKey = vw_silver_factinternetsales_sql_new.ProductKey AND
silver_factinternetsales_sql.OrderDateKey = vw_silver_factinternetsales_sql_new.OrderDateKey AND
silver_factinternetsales_sql.DueDateKey = vw_silver_factinternetsales_sql_new.DueDateKey AND
silver_factinternetsales_sql.ShipDateKey = vw_silver_factinternetsales_sql_new.ShipDateKey AND
silver_factinternetsales_sql.CustomerKey = vw_silver_factinternetsales_sql_new.CustomerKey AND
silver_factinternetsales_sql.PromotionKey = vw_silver_factinternetsales_sql_new.PromotionKey AND
silver_factinternetsales_sql.CurrencyKey = vw_silver_factinternetsales_sql_new.CurrencyKey AND
silver_factinternetsales_sql.SalesTerritoryKey = vw_silver_factinternetsales_sql_new.SalesTerritoryKey

WHEN MATCHED THEN 

UPDATE SET 
silver_factinternetsales_sql.SalesOrderNumber = vw_silver_factinternetsales_sql_new.SalesOrderNumber,
silver_factinternetsales_sql.SalesOrderLineNumber = vw_silver_factinternetsales_sql_new.SalesOrderLineNumber,
silver_factinternetsales_sql.RevisionNumber = vw_silver_factinternetsales_sql_new.RevisionNumber,
silver_factinternetsales_sql.OrderQuantity = vw_silver_factinternetsales_sql_new.OrderQuantity,
silver_factinternetsales_sql.UnitPrice = vw_silver_factinternetsales_sql_new.UnitPrice,
silver_factinternetsales_sql.ExtendedAmount = vw_silver_factinternetsales_sql_new.ExtendedAmount,
silver_factinternetsales_sql.UnitPriceDiscountPct = vw_silver_factinternetsales_sql_new.UnitPriceDiscountPct,
silver_factinternetsales_sql.DiscountAmount = vw_silver_factinternetsales_sql_new.DiscountAmount,
silver_factinternetsales_sql.ProductStandardCost = vw_silver_factinternetsales_sql_new.ProductStandardCost,
silver_factinternetsales_sql.TotalProductCost = vw_silver_factinternetsales_sql_new.TotalProductCost,
silver_factinternetsales_sql.SalesAmount = vw_silver_factinternetsales_sql_new.SalesAmount,
silver_factinternetsales_sql.TaxAmt = vw_silver_factinternetsales_sql_new.TaxAmt,
silver_factinternetsales_sql.Freight = vw_silver_factinternetsales_sql_new.Freight,
silver_factinternetsales_sql.CarrierTrackingNumber = vw_silver_factinternetsales_sql_new.CarrierTrackingNumber,
silver_factinternetsales_sql.CustomerPONumber = vw_silver_factinternetsales_sql_new.CustomerPONumber,
silver_factinternetsales_sql.OrderDate = vw_silver_factinternetsales_sql_new.OrderDate,
silver_factinternetsales_sql.DueDate = vw_silver_factinternetsales_sql_new.DueDate,
silver_factinternetsales_sql.ShipDate = vw_silver_factinternetsales_sql_new.ShipDate


WHEN NOT MATCHED THEN 

INSERT  
(
silver_factinternetsales_sql.ProductKey,
silver_factinternetsales_sql.OrderDateKey,
silver_factinternetsales_sql.DueDateKey,
silver_factinternetsales_sql.ShipDateKey,
silver_factinternetsales_sql.CustomerKey,
silver_factinternetsales_sql.PromotionKey,
silver_factinternetsales_sql.CurrencyKey,
silver_factinternetsales_sql.SalesTerritoryKey,
silver_factinternetsales_sql.SalesOrderNumber,
silver_factinternetsales_sql.SalesOrderLineNumber,
silver_factinternetsales_sql.RevisionNumber,
silver_factinternetsales_sql.OrderQuantity,
silver_factinternetsales_sql.UnitPrice,
silver_factinternetsales_sql.ExtendedAmount,
silver_factinternetsales_sql.UnitPriceDiscountPct,
silver_factinternetsales_sql.DiscountAmount,
silver_factinternetsales_sql.ProductStandardCost,
silver_factinternetsales_sql.TotalProductCost,
silver_factinternetsales_sql.SalesAmount,
silver_factinternetsales_sql.TaxAmt,
silver_factinternetsales_sql.Freight,
silver_factinternetsales_sql.CarrierTrackingNumber,
silver_factinternetsales_sql.CustomerPONumber,
silver_factinternetsales_sql.OrderDate,
silver_factinternetsales_sql.DueDate,
silver_factinternetsales_sql.ShipDate
)
VALUES (
vw_silver_factinternetsales_sql_new.ProductKey,
vw_silver_factinternetsales_sql_new.OrderDateKey,
vw_silver_factinternetsales_sql_new.DueDateKey,
vw_silver_factinternetsales_sql_new.ShipDateKey,
vw_silver_factinternetsales_sql_new.CustomerKey,
vw_silver_factinternetsales_sql_new.PromotionKey,
vw_silver_factinternetsales_sql_new.CurrencyKey,
vw_silver_factinternetsales_sql_new.SalesTerritoryKey,
vw_silver_factinternetsales_sql_new.SalesOrderNumber,
vw_silver_factinternetsales_sql_new.SalesOrderLineNumber,
vw_silver_factinternetsales_sql_new.RevisionNumber,
vw_silver_factinternetsales_sql_new.OrderQuantity,
vw_silver_factinternetsales_sql_new.UnitPrice,
vw_silver_factinternetsales_sql_new.ExtendedAmount,
vw_silver_factinternetsales_sql_new.UnitPriceDiscountPct,
vw_silver_factinternetsales_sql_new.DiscountAmount,
vw_silver_factinternetsales_sql_new.ProductStandardCost,
vw_silver_factinternetsales_sql_new.TotalProductCost,
vw_silver_factinternetsales_sql_new.SalesAmount,
vw_silver_factinternetsales_sql_new.TaxAmt,
vw_silver_factinternetsales_sql_new.Freight,
vw_silver_factinternetsales_sql_new.CarrierTrackingNumber,
vw_silver_factinternetsales_sql_new.CustomerPONumber,
vw_silver_factinternetsales_sql_new.OrderDate,
vw_silver_factinternetsales_sql_new.DueDate,
vw_silver_factinternetsales_sql_new.ShipDate
);
```
